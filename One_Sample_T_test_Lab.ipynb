```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def one_sample_ttest(sample, popmean, alpha):
    # Visualize sample distribution for normality
    sns.histplot(sample, kde=True)
    plt.title('Sample Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

    # Population mean
    pop_mean = popmean

    # Sample mean (x̄)
    sample_mean = np.mean(sample)

    # Sample Standard Deviation (s)
    sample_stdev = np.std(sample, ddof=1)

    # Degrees of freedom
    df = len(sample) - 1

    # Calculate the t-value
    t_statistic = (sample_mean - pop_mean) / (sample_stdev / np.sqrt(len(sample)))

    # Calculate the critical t-value
    t_critical = stats.t.ppf(1 - alpha, df)

    # Calculate the p-value
    p_value = 1 - stats.t.cdf(t_statistic, df)

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"T-statistic: {t_statistic}")
    print(f"Critical T-value: {t_critical}")
    print(f"P-value: {p_value}")

    # Check if the result is significant
    if p_value < alpha:
        print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the sample mean is significantly different from the population mean.")

    return t_statistic, t_critical, p_value

# Example usage:
sample = [434, 694, 457, 534, 720, 400, 484, 478, 610, 641, 425, 636, 454, 514, 563, 370, 499, 640, 501, 625, 612, 471, 598, 509, 531]
popmean = 500
alpha = 0.05

one_sample_ttest(sample, popmean, alpha)
```

```python
# Original scores
sample_original = [84.0, 92.4, 74.3, 79.4, 86.7, 75.3, 90.9, 86.1, 81.0, 85.1,
                    78.7, 73.5, 86.9, 87.4, 82.7, 81.9, 69.9, 77.2, 79.3, 83.3]

# Population mean
popmean = 65
alpha = 0.05

# Perform one-sample t-test
t_statistic, t_critical, p_value = one_sample_ttest(sample_original, popmean, alpha)

# Print results
print(f"Original Scores: T-statistic = {t_statistic}, P-value = {p_value}")
```
